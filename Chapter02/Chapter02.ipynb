```python
# coding: utf-8
# ---------------------
# ---------------------
# Exercise of Chapter 2
# ---------------------
# ---------------------

# ---------------------
# Ex. 2.5: Для каких значений N выполняется неравенство 10NlgN > 2N**2 ?
# ---------------------

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

params = {'legend.fontsize': 14,
          'legend.handlelength': 2}
plt.rcParams.update(params)

# Графический способ

def lg_func(_N):
    """f(N)=lgN"""
    return 10*_N*np.log2(_N)

def sq_func(_N):
    """f(N)=N**2"""
    return 2*_N**2


def plot_func(iterations, first_func, second_func, _ax):
    N = np.array(range(1, iterations))
    first = first_func(N)
    second = second_func(N)

    # plot data        
    _ax.scatter(N, first, color="red", marker="o", label=first_func.__doc__)
    _ax.scatter(N, second, color='blue', marker='x', label=second_func.__doc__)
    _ax.set_xlabel("N")
    _ax.set_ylabel("values")
    _ax.set_title(f'{iterations} iterations')
    _ax.legend(loc=2)
    _ax.grid(True)


fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
## N = [1, 25]
plot_func(25, lg_func, sq_func, _ax=ax[0])

## N = [1, 100]
plot_func(100, lg_func, sq_func, _ax=ax[1])

plt.show()

# как видим, уже после 23 итерации квадратичная функция больше логарифмической.
# Ответ: 10NlgN > 2N**2 при N = [1..22]
```
